```python
import pandas as pd
from pandas import Series, DataFrame
import os
```

```python
path = 'cleaned data/'
clean_files = os.listdir(path)
```

```python
csv = []
for file in clean_files:
    if '.csv' in file:
        name = path + file
        csv.append(name)
```

```python
frame = pd.read_csv('cleaned data/frame.csv')
frame
```

```python
df_list = [frame]
for file in csv:
    file = pd.read_csv(file)
    df_list.append(file)
```

```python
all_tech_df = pd.concat(df_list)
all_tech_df
```
