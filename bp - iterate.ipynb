```python
import pandas as pd
from pandas import Series, DataFrame
```

```python
missing_values = ['-', '^','♦']
bp = pd.read_excel('bp-stats-review-2022-all-data.xlsx', sheet_name='Nuclear Generation - TWh', header=2, index_col=0, na_values=missing_values)
bp
```

```python
bp = bp.dropna(how='all')
bp.index.name = 'Country Name'
bp
```

```python
for country in bp.index:
     if 'Total' in country or 'Other' in country or 'European Union' in country or 'OECD' in country or 'Central America' in country or 'Eastern Africa' in country or 'Middle Africa' in country or 'Western Africa' in country:
            bp = bp.drop(country)
bp
```

```python
bp = bp.dropna(how='all')
bp = bp.drop(columns=['2021.1', '2011-21', '2021.2'])
bp
```

```python
iso = pd.read_excel('WUP2018-DataSource-UrbanAgglomeration-and-CapitalCities.xls', header=16, usecols=[1, 3, 4])
iso = iso.drop_duplicates()
iso = iso.set_index('Country or area')
```

```python
iso2 = []
err = []
for country in bp.index:
    try:
        iso2.append(iso.loc[country][1])
    except KeyError:
        if country=='US':
            iso2.append(iso.loc['United States of America'][1])
        elif country=='Venezuela':
            iso2.append(iso.loc['Venezuela (Bolivarian Republic of)'][1])
        elif country=='Czech Republic':
            iso2.append(iso.loc['Czechia'][1])
        elif country=='South Korea':
            iso2.append(iso.loc['Republic of Korea'][1])
        elif country=='Vietnam':
            iso2.append(iso.loc['Viet Nam'][1])
        elif country=='USSR':
            iso2.append('SU')
        elif country=='Trinidad & Tobago':
            iso2.append(iso.loc['Trinidad and Tobago'][1])
        elif country=='North Macedonia':
            iso2.append(iso.loc['TFYR Macedonia'][1])
        elif country=='Iran':
            iso2.append(iso.loc['Iran (Islamic Republic of)'][1])
        elif country=='Taiwan':
            iso2.append(iso.loc['China, Taiwan Province of China'][1])
        elif country=='China Hong Kong SAR':
            iso2.append(iso.loc['China, Hong Kong SAR'][1])
        else:
            err.append(country)

err
```

```python
if len(err) != 0:
    raise Exception('Something went wrong')

bp['Country Code'] = iso2
```

```python
bp.insert(0, 'Country Name', bp.index)
bp = bp.set_index('Country Code')
bp
```
