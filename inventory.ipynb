```python
import pandas as pd
from pandas import Series, DataFrame
```

```python
all_tech = pd.read_csv('all_tech.csv')
all_tech
```

```python
tech_list = sorted(list(set(all_tech['Technology Name'])))
len(tech_list)
```

```python
del_list = ['Automobile', 'Biofuels', 'Cellular phone','Coal Power','Dryer','Electric power',
                 'Electricity access','Ethylene 2','Household refrigerator','Hydro Power','Hydroelectricity',
                 'Iron','Landlines','NatGas Power','Nuclear Power','On-Grid Coal And Peat','On-Grid Natural Gas',
                 'On-Grid Nuclear','On-Grid Oil','Passenger Cars','Primary Alumninum','Refrigerators',
                 'Solar Energy','Solar Power','Steamships','Washer','Washing machine','Wind Energy','Wind Power']
for item in del_list:
    tech_list.remove(item)
len(tech_list)
                 
```

```python
tech_us_global = {}
for tech in tech_list:
    country_names = list(set(all_tech[all_tech['Technology Name'] == tech]['Country Name']))
    country_codes = list(set(all_tech[all_tech['Technology Name'] == tech]['Country Code']))
    if 'World' in country_names or 'US' in country_codes:
        tech_us_global[tech] = 'x'
    else:
        tech_us_global[tech] = ''
        
for entry in tech_us_global:
    if tech_us_global[entry] == '':
        print(entry)
```

```python
tech_country_dict = {}
for tech in tech_list:
    countries = list(set(all_tech[all_tech['Technology Name'] == tech]['Country Name']))
    if len(countries) > 2:
        tech_country_dict[tech] = len(countries)
    else:
        tech_country_dict[tech] = countries      
```

```python
for val in tech_country_dict:
    if type(tech_country_dict[val]) != int:
        print(val, tech_country_dict[val])
```

```python
multi_list = []
for val in tech_country_dict:
    if type(tech_country_dict[val]) == int:
        multi_list.append(val)
len(multi_list)
```
