```python
import pandas as pd
from pandas import Series, DataFrame
```

```python
all_tech = pd.read_csv('all_tech.csv')
all_tech
```

```python
tech_list = sorted(list(set(all_tech['Technology Name'])))
len(tech_list)
```

```python
tech_country_dict = {}
for tech in tech_list:
    tech_country_dict[tech] = len(set(all_tech[all_tech['Technology Name'] == tech]['Country Name']))
```

```python
tech_country_dict = dict(sorted(tech_country_dict.items(), key = lambda item: item[1]))
tech_country_dict
```

```python
one_country_list = []
mult_country_list = []
for tech in tech_country_dict:
    val = tech_country_dict[tech]
    if val == 1:
        one_country_list.append(tech)
    else:
        mult_country_list.append(tech)
print(len(one_country_list))
print(len(mult_country_list))
```

```python
country_lt_5 = []
country_lt_50 = []
country_lt_250 = []
for tech in tech_country_dict:
    val = tech_country_dict[tech]
    if val > 1 and val < 5:
        country_lt_5.append(tech)
    if val > 4 and val < 50:
        country_lt_50.append(tech)
    if val > 49 and val < 250:
        country_lt_250.append(tech)
print(len(country_lt_5))
print(len(country_lt_50))
print(len(country_lt_250))
```
