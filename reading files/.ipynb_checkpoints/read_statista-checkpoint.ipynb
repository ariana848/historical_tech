```python
# Does Jenna have access to see the underlying data source?
satellites = pd.read_csv('raw data/Satellite Launches.csv', index_col=0)
satellites = satellites.transpose()
satellites['Data Source'] = 'Statista'
satellites['Country Name'] = 'World'
satellites['Country Code'] = 'World'
satellites['Spatial Scale'] = 'Global'
satellites['Metric'] = 'Cumulative Total Capacity'
satellites['Technology Name'] = 'Satellite launches'
satellites['Unit'] = 'Number of orbital launches'
satellites['ID'] = satellites['Technology Name'] + '_' + satellites['Metric'] + '_' + satellites['Country Code']
satellites.set_index('ID', drop=True, inplace=True)
satellites.columns.name = None
satellites
```
