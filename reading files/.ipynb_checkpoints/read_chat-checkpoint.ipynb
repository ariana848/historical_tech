```python
ag_harvester = pd.read_csv('raw data/FinalCHAT_72909.csv', usecols=[0,1,2])
ag_harvester
# see read_doe, same data set up, use that code!
```

```python
country_list = sorted(list(set(ag_harvester['country_name'])))
len(country_list)
```

```python
## write a function that takes the var_name and then spits out the df that looks like the below (for all countries)
## must give var_name and tech name, unit, metric as arguments
## will have to go back and take out empty rows since even countries/years without data are included as rows (see IRENA?)

ag_harvester_alb = ag_harvester[ag_harvester['country_name']==country_list[1]]
ag_harvester_alb.set_index('year', drop=True, inplace=True)
ag_harvester_alb = ag_harvester_alb.transpose()
ag_harvester_alb['Country Name'] = country_list[1]
ag_harvester_alb['Country Code'] = coco.convert(names=country_list[1], to='iso2')
ag_harvester_alb.drop(index='country_name', inplace=True)
ag_harvester_alb['Data Source'] = 'NBER CHAT'
ag_harvester_alb['Spatial Scale'] = 'National'
ag_harvester_alb['Unit'] = 'Number in operation'
ag_harvester_alb['Metric'] = 'Cumulative total capacity'
ag_harvester_alb['Technology Name'] = 'Crop harvester'
ag_harvester_alb['ID'] = ag_harvester_alb['Technology Name'] + '_' + ag_harvester_alb['Metric'] + '_' + ag_harvester_alb['Country Code']
ag_harvester_alb.set_index('ID', drop=True, inplace=True)
ag_harvester_alb.columns.name = None
ag_harvester_alb
```
