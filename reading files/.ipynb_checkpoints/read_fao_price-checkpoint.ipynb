```python
# corn
corn = pd.read_csv('raw data/cost data/FAOSTAT_data_en_7-7-2023.csv')
corn = corn[corn['Item']=='Maize (corn)']
corn = corn.pivot(index='Area', columns='Year', values='Value')
corn['Unit'] = 'Producer Price Index (2014-2016 = 100)'
corn['Metric'] = 'Price'
corn['Spatial Scale'] = 'National'
corn['Technology Name'] = 'Corn Production'
corn['Data Source'] = 'FAOSTAT'
corn['Country Name'] = corn.index
iso = []
for country in corn.index:
    iso2 = coco.convert(names=country, to='iso2')
    iso.append(iso2)
corn['Country Code'] = iso
corn['ID'] = corn['Technology Name'] + '_' + corn['Metric'] + '_' + corn['Country Code']
corn.set_index('ID', drop=True, inplace=True)
corn.columns.name=None
corn.to_csv('cleaned data/corn_prices.csv')
corn
```
