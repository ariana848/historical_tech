```python
nass_soy = pd.read_csv('raw data/D56D3FF7-028A-31E2-AA66-8A643B839228.csv', index_col=0, usecols = [1,19])
nass_soy = nass_soy.transpose()
nass_soy['name'] = ['soy - acres planted']
nass_soy.set_index('name', inplace=True)
nass_soy.columns.name = None
nass.append(nass_soy)
```
