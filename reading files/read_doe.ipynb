```python
tech_list = list(set(doe['Technology Mid-Type']))
tech_list.sort()
for x in tech_list:
    print(x)
```

```python
doe.set_index('Year of Project Year', drop=True, inplace=True)
doe
```

```python
def read_doe(tech_name):
    df = doe[doe['Technology Mid-Type'] == tech_name]
    country_list = list(set(df['Country']))
    df_list = []
    for country in country_list:
        df_country = df[df['Country']== country]
        df_country = df_country.transpose()
        df_country['Technology Name'] = tech_name
        df_country['Country Name'] = country
        df_country.reset_index(inplace=True, drop=True)
        df_country = df_country[2:]
        df_country['Unit'] = 'kW'
        df_country['Data Source'] = 'DOE Global Energy Storage Database'
        df_country['Metric'] = 'Cumulative total capacity'
        df_country['Spatial Scale'] = 'National'
        df_country['Country Code'] = coco.convert(names=country, to='iso2')
        df_country['ID'] = df_country['Technology Name'] + '_' + df_country['Metric'] + '_' + df_country['Country Code']
        df_country.set_index('ID', drop=True, inplace=True)
        df_country.columns.name = None
        df_list.append(df_country)
        file_name = 'doe/' + tech_name + country + '.csv'
        df_country.to_csv(file_name)
    return df_list
```

```python
for tech in tech_list:
    read_doe(tech)
```
