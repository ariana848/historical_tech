```python
import pandas as pd
from pandas import Series, DataFrame
import country_converter as coco
```

```python
def read_fao(file_name):
    df = pd.read_excel(file_name)
    df = df.drop(0)
    df = df.drop(columns=['Unnamed: 1', 'Unnamed: 2'])
    df = df.set_index('Unnamed: 0')
    df.index.name = 'Country Name'
    df = df.dropna(how='all')
    iso2 = []
    for country in df.index:
        if country=='China, Taiwan Province of':
            iso2.append('TW')
        elif country=='Serbia and Montenegro':
            iso2.append('CS')
        elif country=='USSR':
            iso2.append('SU')
        elif country=='Yugoslav SFR':
            iso2.append('YU')
        elif country=='Belgium-Luxembourg':
            iso2.append('BE-WLX')   
        else:
            iso2.append(coco.convert(names=country, to='iso2'))
    df['Country Code'] = iso2
    df.insert(0, 'Country Name', df.index)
    df = df.set_index('Country Code')
    df.insert(1, 'Unit', 'Tonnes')
    return df
```

```python
fao_nitrogen = read_fao('FAOSTAT_data_2-9-2023.xls')
fao_nitrogen.insert(1, 'Technology Name', 'Nitrogen fertilizer production')
fao_nitrogen
```
