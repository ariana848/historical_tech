```python
nipa = pd.read_excel('A001RG3A086NBEA.xls', header=10)
nipa
```

```python
year_list = []
for x in nipa['observation_date']:
    x = int(str(x)[:4])
    year_list.append(x)
nipa['Year'] = year_list
nipa.set_index('Year', drop=True, inplace=True)
nipa.drop(columns='observation_date', inplace=True)
nipa = nipa.transpose()
nipa
```

```python
all_tech = pd.read_csv('all_tech.csv')
all_tech
```

```python
cost_data = all_tech[all_tech['Metric']=='Price']
cost_data.set_index('ID', drop=True, inplace=True)
cost_data.dropna(axis=1, how='all', inplace=True)
cost_data
```

```python
usgs = cost_data[cost_data['Data Source']=='U.S. Geological Survey']
usgs
```
